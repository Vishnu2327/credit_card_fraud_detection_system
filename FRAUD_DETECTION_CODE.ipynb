```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, accuracy_score



csv_path = "/content/creditcard.csv"
df = pd.read_csv(csv_path)


features_to_keep = df.columns[1:5] 
X = df[features_to_keep]
y = df['Class']


fraud_idx = y[y == 1].index
nonfraud_idx = y[y == 0].index
sample_size = min(5000, len(nonfraud_idx))
nonfraud_sample_idx = np.random.choice(nonfraud_idx, size=sample_size, replace=False)
sample_idx = np.concatenate([fraud_idx, nonfraud_sample_idx])
X_sampled = X.loc[sample_idx]
y_sampled = y.loc[sample_idx]


X_train, X_test, y_train, y_test = train_test_split(
    X_sampled, y_sampled, test_size=0.2, stratify=y_sampled, random_state=42
)


scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)


lr_model = LogisticRegression(
    max_iter=200,
    C=0.05,
    class_weight='balanced',
    random_state=42
)
lr_model.fit(X_train_scaled, y_train)
y_pred_lr = lr_model.predict(X_test_scaled)
y_prob_lr = lr_model.predict_proba(X_test_scaled)[:,1]


rf_model = RandomForestClassifier(
    n_estimators=10,
    max_depth=3,
    max_features='log2',
    class_weight='balanced',
    random_state=42
)
rf_model.fit(X_train_scaled, y_train)
y_pred_rf = rf_model.predict(X_test_scaled)
y_prob_rf = rf_model.predict_proba(X_test_scaled)[:,1]


scale_pos_weight = (y_train == 0).sum() / (y_train == 1).sum()

xgb_model = XGBClassifier(
    n_estimators=25,      
    max_depth=4,          
    learning_rate=0.05,
    scale_pos_weight=scale_pos_weight,
    use_label_encoder=False,
    eval_metric='logloss',
    random_state=42
)
xgb_model.fit(X_train_scaled, y_train)
y_pred_xgb = xgb_model.predict(X_test_scaled)
y_prob_xgb = xgb_model.predict_proba(X_test_scaled)[:,1]



def evaluate_model(name, y_true, y_pred, y_prob):
    print(f"===== {name} =====")
    print("Confusion Matrix:\n", confusion_matrix(y_true, y_pred))
    print("\nClassification Report:\n", classification_report(y_true, y_pred, digits=4))
    print(f"Accuracy: {accuracy_score(y_true, y_pred):.4f}")
    print(f"ROC-AUC Score: {roc_auc_score(y_true, y_prob):.4f}\n")


evaluate_model("Logistic Regression", y_test, y_pred_lr, y_prob_lr)
evaluate_model("Random Forest", y_test, y_pred_rf, y_prob_rf)
evaluate_model("XGBoost", y_test, y_pred_xgb, y_prob_xgb)
```

```python
import pandas as pd
import torch
import torch.nn.functional as F
import numpy as np
from torch_geometric.data import Data
from torch_geometric.nn import SAGEConv
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score, accuracy_score
from sklearn.preprocessing import StandardScaler
import random

df = pd.read_csv("/content/creditcard.csv")


df_fraud = df[df['Class']==1].sample(n=400, random_state=42)
df_nonfraud = df[df['Class']==0].sample(n=19600, random_state=42)
df_small = pd.concat([df_fraud, df_nonfraud]).sample(frac=1, random_state=42).reset_index(drop=True)


features = df_small.drop(columns=['Class']).values
labels = df_small['Class'].values
features = StandardScaler().fit_transform(features)

x = torch.tensor(features, dtype=torch.float)
y = torch.tensor(labels, dtype=torch.long)
num_nodes = x.shape[0]


time_bucket = (df_small['Time'] // 1000).astype(int)
edges = []
bucket_dict = {}
for idx, b in enumerate(time_bucket):
    bucket_dict.setdefault(b, []).append(idx)

for nodes in bucket_dict.values():
    for i in range(len(nodes)):
        for j in range(i+1, len(nodes)):
            edges.append([nodes[i], nodes[j]])
            edges.append([nodes[j], nodes[i]])


num_random_edges = int(0.05 * len(edges))
for _ in range(num_random_edges):
    a, b = random.randint(0, num_nodes-1), random.randint(0, num_nodes-1)
    edges.append([a, b])
    edges.append([b, a])

edge_index = torch.tensor(edges, dtype=torch.long).t().contiguous()
data = Data(x=x, edge_index=edge_index, y=y)


train_idx, test_idx = train_test_split(range(num_nodes), test_size=0.2, random_state=42, stratify=y)

class GraphSAGE(torch.nn.Module):
    def __init__(self, in_channels, hidden_channels, out_channels):
        super().__init__()
        self.conv1 = SAGEConv(in_channels, hidden_channels)
        self.conv2 = SAGEConv(hidden_channels, hidden_channels)
        self.conv3 = SAGEConv(hidden_channels, out_channels)

    def forward(self, data):
        x, edge_index = data.x, data.edge_index
        x = F.relu(self.conv1(x, edge_index))
        x = F.dropout(x, p=0.4, training=self.training)
        x = F.relu(self.conv2(x, edge_index))
        x = F.dropout(x, p=0.4, training=self.training)
        x = self.conv3(x, edge_index)
        return x


model = GraphSAGE(x.shape[1], 256, 2)
optimizer = torch.optim.Adam(model.parameters(), lr=0.001, weight_decay=1e-4)


num_class0 = (y==0).sum().item()
num_class1 = (y==1).sum().item()
weight = torch.tensor([1.0, num_class0/num_class1], dtype=torch.float)
criterion = torch.nn.CrossEntropyLoss(weight=weight)


model.train()
optimizer.zero_grad()
out = model(data)
y_train_tensor = y[train_idx].clone().detach()  
loss = criterion(out[train_idx], y_train_tensor)
loss.backward()
optimizer.step()


model.eval()
out = model(data)
y_prob_tensor = F.softmax(out[test_idx], dim=1)[:,1].detach().numpy()


custom_thresh = 0.95  
y_pred_tensor = (y_prob_tensor >= custom_thresh).astype(int)

acc = accuracy_score(y[test_idx].numpy(), y_pred_tensor)
cm = confusion_matrix(y[test_idx].numpy(), y_pred_tensor)
cr = classification_report(y[test_idx].numpy(), y_pred_tensor, zero_division=0)
roc_auc = roc_auc_score(y[test_idx].numpy(), y_prob_tensor)

print(f"Test Accuracy: {acc:.4f}")
print("Confusion Matrix:\n", cm)
print("\nClassification Report:\n", cr)
print(f"ROC-AUC Score: {roc_auc:.4f}")


```
